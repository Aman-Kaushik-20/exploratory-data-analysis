```python
import numpy as np
import pandas as pd
import seaborn as sns
df=pd.read_csv('/content/drive/MyDrive/Datasets/Black Friday Sale/BlackFriday.csv')
del df['Product_Category_2']
del df['Product_Category_3']
df.head()
```

```python
l=[]

for i in range(len(df)):
  l.append(df['Gender'][i]+'_'+str(df['Marital_Status'][i]))

df['Gender__Marital_Status']=l
```
