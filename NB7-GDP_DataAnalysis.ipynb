```python
import numpy as np
import pandas as pd
import seaborn as sns

df=pd.read_csv('/content/drive/MyDrive/Datasets/GDP Analysis/gdp.csv')

df.head()
```

```python
df_pr=df[df['Country Name']=='India']

data=df_pr.values

gdp_change=[0]

for i in range(1,len(data)):

  prev=data[i-1][3]
  curr=data[i][3]
  growth=( ( curr - prev ) / prev)*100

  gdp_change.append( round(growth,2) )

df_pr=df_pr.assign(GDP_Growth=gdp_change)

df_pr.head()
```

```python
df_pr=df[df['Country Name']=='China']

data=df_pr.values

gdp_change=[0]

for i in range(1,len(data)):

  prev=data[i-1][3]
  curr=data[i][3]
  growth=( ( curr - prev ) / prev)*100

  gdp_change.append( round(growth,2) )

df_pr=df_pr.assign(GDP_Growth=gdp_change)

df_pr.head()
```

```python
import numpy as np
import pandas as pd
import plotly.express as px
import plotly.offline as pyo
import os

df=pd.read_csv('/content/drive/MyDrive/Datasets/GDP Analysis/gdp.csv')

final_lst=[]
for name in df['Country Name'].unique():

  df_pr=df[df['Country Name']==name]
  data=df_pr.values
  gdp_change=[0]

  for i in range(1,len(data)):

    prev=data[i-1][3]
    curr=data[i][3]
    growth=( ( curr - prev ) / prev)*100

    gdp_change.append( round(growth,2) )

  df_pr=df_pr.assign(GDP_Growth=gdp_change)
  final_lst.append(df_pr)

df=pd.concat(final_lst,axis=0)
df.head()
```
